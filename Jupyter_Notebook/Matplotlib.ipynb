```python
def displayFigure(data, xlim=[0,0], ylim=[0,0], tick=[0,0], title=""):
    def findDepth(orgList):
        l = orgList.copy()
        cpt = 0

        while isinstance(l, (list, np.ndarray)):
            l = l[0]
            cpt += 1
        return cpt


    fig, ax = plt.subplots()
    ax.set_title(title)
    
    plt.grid(True, linestyle=':')
    
    if xlim != [0,0]:
        ax.set_xlim(xlim[0], xlim[1])
        ax.set_xticks(np.arange(xlim[0], xlim[1], tick[0]))
    
    if ylim != [0,0]:      
        ax.set_ylim(ylim[0], ylim[1])
        ax.set_yticks(np.arange(ylim[0], ylim[1], tick[1]))
    
    ax.tick_params(which='major', direction='inout', length=5, width=2, colors='k')
    ax.tick_params(which='minor', direction='out', length=3, width=1, colors='k')
    
    ax.spines['bottom'].set_position(('data', 0))
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_position(('data', 0))
    
    nbPlot = findDepth(data) - 1
    if nbPlot == 1:
        ax.plot(data[0], data[1])
    elif nbPlot >= 2:
        for p in data:
            #ax.scatter(p[0], p[1])
            ax.plot(p[0], p[1])
        
    
    plt.show()
    
    return None
```

```python
displayFigure(curve2, xlim=[-3, 7], tick=[1, 0.25], title='Curve')
```
